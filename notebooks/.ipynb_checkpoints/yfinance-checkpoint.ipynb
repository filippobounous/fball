```python
from private.consts.file_paths import HISTORICAL_DATA_PATH

def prepare_yf_ticker_data(ticker_code='0P0000OQVL.F'):
    ticker = yf.Ticker(ticker_code)
    df = ticker.history(
        start=datetime.datetime(2025,1,1),
        end=datetime.date.today()
    )
    df.index.name = "as_of_date"
    
    df.columns = (
        df.columns
        .str.strip()            # Remove leading/trailing spaces
        .str.lower()             # Convert to lowercase
        .str.replace(' ', '_')   # Replace spaces with underscores
        .str.replace(r'[^\w_]', '', regex=True)  # Remove non-word characters
    )
    
    df.index = pd.to_datetime(df.index)
    df.index = df.index.tz_localize(None)
    df.index = df.index.normalize()
    
    df.to_csv(f"{HISTORICAL_DATA_PATH}/yahoo_finance/{ticker_code}.csv")
    return df
```

```python
prepare_yf_ticker_data()
```
